```python
#required imports
import numpy as np
import pandas as pd
import cvxopt as cp
import math
```

QUESTION 1

```python
#QP
Q = np.array([
        [2, 1, 0], 
        [1, 2, 1], 
        [0, 1, 2]
    ])

p = np.array([0, 0, 0])
h = np.array([[-0.25], [0], [0], [0]])
G = np.array([
    [-0.4, -0.4, -0.8],
    [-1, 0, 0],
    [0, -1, 0],
    [0, 0, -1]
])
A_eq = np.array([
        [1, 1, 1]
])
b_eq = np.array(
    [
    [1]
    ]
)

result = cp.solvers.qp(
    cp.matrix(Q, tc='d'),
    cp.matrix(p, tc='d'),
    cp.matrix(G, tc='d'),
    cp.matrix(h, tc='d'),
    cp.matrix(A_eq, tc='d'),
    cp.matrix(b_eq, tc='d')
)

print("Answer:")
print(result['x'], result['primal objective'])
print(result["primal objective"])
```
