```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('2 columns.xls', header = None)
data.columns = ['x', 'y']
data.head()
```

```python
B = data.values
B
```

```python
x = B[:,0]
y = B[:,1]
```

### Degree = 2

```python
A = np.column_stack((x**2, x, np.ones((len(x), 1), dtype = float)))
A[:5]
beta = np.dot(np.linalg.inv(np.dot(A.T, A)), np.dot(A.T, y))
beta

plt.figure(figsize = (12, 8))
x = np.sort(x)
plt.plot(x, y, 'o', label = 'data')
plt.plot(x, beta[0]*x**2 + beta[1]*x + beta[2], label = 'fit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

val = 2000; #Roll Number - B20AI052
y = beta[0]*val**2 + beta[1]*val + beta[2]
print("The predicted value for R = 2000 with degree 2 is:",  y)
```
